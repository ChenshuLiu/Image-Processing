# Digital Images

Digital images are just matrices of numbers

Digital image processing is the process of manipulation of the numbers that give rise to the image attributes thus rendering different visual effects of the original image

```python
import cv2
from skimage import io
import numpy as np
from matplotlib import pyplot as plt
img = io.imread("image.jpg")
plt.imshow(img)
# adding a red box to the image
img[10:200, 10:200, :] = [255, 0, 0]
plt.imshow(img)
```

# Image processing using pillow

```python
from PIL import Image # load images
from matplotlib import pyplot as plt
img = Image.open("image.jpg")
# NOTE: when images are opened in pillow (i.e. PIL) the type is not np array!
print("data type is:", type(img))
print("mode of image is:", img.mode)
print("format of the image is:", img.format)
print("image size is (in width x height):", img.size)
```

```python
# resizing by squeezing
small_img = img.resize((100, 300))
# saving the resized image
# small_img.save("path/new_image_name")
# resizing just simply squeeze the image into the desired dimension
plt.imshow(small_img)
```

```python
# resizing while keeping aspect ratio
img.thumbnail((100, 300))
plt.imshow(img)
print(img.size)
# NOTE: the resulting shape of the image is not as what is designated, this is because the aspect ratio is kept
```
